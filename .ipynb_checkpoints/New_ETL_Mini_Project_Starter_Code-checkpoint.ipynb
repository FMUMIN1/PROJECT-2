```python
# Import the dependencies
import pandas as pd
from pprint import pprint
from pymongo import MongoClient
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
```

# NETFLIX

```python
# load csv file
netflixDF = pd.read_csv('NETFLIX/titles.csv')
netflixDF.head()
```

```python
#select desired columns
netflixDF = netflixDF[['id','title','type','release_year', 'genres','production_countries', 'imdb_score','imdb_votes']]

# drop missing values 
netflixDF = netflixDF.dropna()

# order rows based on IMDB SCORES AND VOTES, in ascending order 
netflixDF = netflixDF.sort_values(['imdb_score', 'imdb_votes'],
              ascending = [False, False])

# reset index
netflixDF = netflixDF.reset_index(drop=True)

netflixDF
```

```python
netflixDF.info()
```

# PARAMOUNT

```python
# load csv file
paramountDF = pd.read_csv('PARAMOUNT/titles.csv')
paramountDF.head()
```

```python
# select desired columns
paramountDF = paramountDF[['id','title','type','release_year', 'genres','production_countries', 'imdb_score','imdb_votes']]

# drop missing values 
paramountDF = paramountDF.dropna()

# order rows based on IMDB SCORES AND VOTES, in ascending order 
paramountDF = paramountDF.sort_values(['imdb_score', 'imdb_votes'],
              ascending = [False, False])

# reset index
paramountDF = paramountDF.reset_index(drop=True)

paramountDF.head()
```

```python
paramountDF.info()
```

# LOAD INTO MONGODB

```python
# create a dictionary from netflix data
netflixDict = netflixDF.to_dict('records')
netflixDict[0]
```

```python
# create a dictionary from paramount data
paramountDict = paramountDF.to_dict('records')
paramountDict[0]
```

```python
# make instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# make a reference to a DB named 'project2DB' with two collections: 'netflix' and 'paramount'
db = mongo.project2DB
netflix = db.netflix
paramount = db.paramount
```

```python
# insert dictionary data into netflix collection within project2DB database
netflix.insert_many(netflixDict)
```

```python
# insert dictionary data into paramount collection within project2DB database
paramount.insert_many(paramountDict)
```

```python
# check to see if the netflix data was loaded
netflixResults = list(netflix.find().limit(3))
for result in netflixResults:
    pprint(result)
```

```python
# check to see if the paramount data was loaded
paramountResults = list(paramount.find().limit(3))
for result in paramountResults:
    pprint(result)
```
